# Data Set

```python
mean_01 = np.array([3.0 ,4.0])
cov_01 =  np.array([[1.0 , -0.5],[-0.5,1.0]])
dist_01 = np.random.multivariate_normal(mean_01 , cov_01,200)
print(dist_01)
```

```python
mean_02 = np.array([0.0 ,0.0])
cov_02 =  np.array([[1.0 ,.5],[0.5,0.6]])
dist_02 = np.random.multivariate_normal(mean_02 , cov_02,200)
print(dist_02)
```

```python
for x in range(dist_01.shape[0]):
    plt.scatter(dist_01[x,0] , dist_01[x,1], color='blue')
    plt.scatter(dist_02[x,0] , dist_02[x,1], color='red') 
    
plt.show()    
```

```python
labels = np.zeros((400,1))
labels[200:] = 1.0
print(labels.shape)
#print(labels)
```

```python
X_data = np.zeros((400 ,2))
X_data[:200,:] = dist_01
X_data[200:,:] = dist_02
```

# Apply KNN Algoritham

```python
def dist(x1 , x2):
    return np.sqrt(((x1-x2)**2).sum())
```

```python
def KNN( x_train , q_pint , y_train , k =5 ):
    vals = list()
    
    for x in range(x_train.shape[0]):
        v = [dist(q_pint ,x_train[x ,:] ) , y_train[x]]
        vals.append(v)
        
    update_val = sorted(vals)
   
    
    pared_arr = np.asarray(update_val[:k])
    print(pared_arr)
    
    label_idx = np.unique(pared_arr[:,1] , return_counts= True)
    
    index = label_idx[1].argmax()
    return label_idx[0][index]
        
       
```

```python
q = np.array([0.0 , 4.0])
lb = predicted_label = KNN(X_data , q, labels)
print(lb)
```
